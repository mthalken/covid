```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read CSV
df = pd.read_csv('data/COVID-19-geo.csv')
df
```

```python
# Look at data
country_list = pd.DataFrame(pd.unique(df['countriesAndTerritories']))
country_list
```

```python
total_deaths = df[['deaths']].sum()
total_cases = df[['cases']].sum()
print(total_deaths)
print(total_cases)
```

```python
df.columns
```

```python
# Group by country and Date
df.groupby(['countriesAndTerritories', 'month'])[['deaths']]
```

```python
new_df = pd.DataFrame(df[['countriesAndTerritories', 'day', 'month', 'year', 'cases', 'deaths']])
new_df
```

```python
clean_df = new_df.groupby(['countriesAndTerritories', 'year', 'month', 'day']).sum()
clean_df
```

```python
total_deaths = clean_df[['deaths']].sum()
total_cases = clean_df[['cases']].sum()
print(total_deaths)
print(total_cases)
```
